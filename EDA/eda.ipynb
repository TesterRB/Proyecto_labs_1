# **Análisis Exploratorio de Datos**

### **Importacion de librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Cargar el archivo CSV y obtener información general del DataFrame**

```python
# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv('dfListo.csv')

# Obtener información general sobre el DataFrame
print(df.info())
```

Este código carga el archivo CSV en un DataFrame de pandas y muestra información general sobre el DataFrame, como el número de filas y columnas, el tipo de datos de cada columna y el número de valores no nulos.

### **Ver las primeras filas del DataFrame**

```python
df.head()
```

Este código muestra las primeras cinco filas del DataFrame para tener una idea de cómo se ven los datos.

### **Obtener estadísticas descriptivas de las columnas numéricas**

```python
print(df.describe())
```

Este código muestra estadísticas descriptivas de las columnas numéricas del DataFrame, como la media, la mediana, la desviación estándar, el valor mínimo y máximo y los percentiles.

### **Grafico de valores nulos dentro delDataFrame**

```python
null_counts = df.isnull().sum()

# Crear el gráfico de barras
plt.figure(figsize=(12,8))
plt.bar(null_counts.index, null_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Columnas')
plt.ylabel('Cantidad de valores nulos')
plt.title('Cantidad de valores nulos por columna')
plt.show()
```

### **Ver la distribución de las variables numéricas**

```python
df.hist(bins=50, figsize=(20,15))
plt.show()
```

En este caso el código muestra histogramas de las variables numéricas del DataFrame para visualizar la distribución de los datos.

### **Ver la correlación entre las variables numéricas**

```python
# Se calcula la matriz de correlación del dataframe y se guarda en la variable corr_matrix.
corr_matrix = df.corr()
# Se utiliza la función heatmap de seaborn para visualizar la matriz de correlación en un mapa de calor
# con anotaciones de los valores en cada celda.
sns.heatmap(corr_matrix, annot=True)
# Se utiliza la función show de matplotlib para mostrar el mapa de calor.
plt.show()
```

Este código muestra una matriz de correlación entre las variables numéricas del DataFrame utilizando un mapa de calor.

### **Ver la distribución de las variables categóricas**

```python
for col in df.select_dtypes(include=['object']):
    print(df[col].value_counts())
```

Este código muestra la distribución de las variables categóricas del DataFrame, contando el número de veces que aparece cada valor en cada columna categórica.

### **Cantidad de peliculas en su respectivo idioma tomadas de un df_sample de 1000 filas**

```python
df_sample = df.sample(n=1000, random_state=42) # Tomar una muestra aleatoria de 1000 filas

# Contar los idiomas en el conjunto de datos
idiomas = df_sample["original_language"].value_counts()

# Visualizar los resultados con un gráfico de barras
idiomas.plot(kind="bar")
plt.title("Películas por idioma original")
plt.xlabel("Idioma")
plt.ylabel("Número de películas")
plt.show()
```

### **Longitud Promedio de los resúmenes**

```python
df["longitud_resumen"] = df["overview"].str.len()
longitud_promedio = df["longitud_resumen"].mean()
print("La longitud promedio de los resúmenes es:", longitud_promedio)
```

```python
# Contar cuántas películas tienen un eslogan y cuántas no
num_eslogans = df["tagline"].count()
num_sin_eslogan = len(df) - num_eslogans

# Imprimir los resultados
print("Hay", num_eslogans, "películas con eslogan.")
print("Hay", num_sin_eslogan, "películas sin eslogan.")
```

```python
# Separar los diferentes géneros y contar cuántas películas están en cada género
generos = df["name_genres"].str.split(",", expand=True).stack().str.strip().value_counts()

# Visualizar los resultados con un gráfico de barras
generos.plot(kind="bar")
plt.title("Películas por género")
plt.xlabel("Género")
plt.ylabel("Número de películas")
plt.show()
```

```python
df_sample = df.sample(n=25, random_state=42) # Tomar una muestra aleatoria de 1000 filas

# Contar cuántas películas produjo cada estudio
estudios = df_sample["name_production"].str.split(",", expand=True).stack().str.strip().value_counts()

# Visualizar los resultados con un gráfico de barras
estudios.plot(kind="bar")
plt.title("Películas por estudio de producción")
plt.xlabel("Estudio de producción")
plt.ylabel("Número de películas")
plt.show()
```
